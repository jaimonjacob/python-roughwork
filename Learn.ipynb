Loading the dataset from google drive

```python
# Reading
# Importing google drive
from google.colab import drive
drive.mount('/content/drive')

# Assigning the file to be read to a variable
file_name = '/content/drive/MyDrive/Learn/lorem.txt'

# Opening the file
with open(file_name) as file_object:

  # reading each line
  lines = file_object.readlines()

new_string = ''

# Adding the content in the text file to the string variable
for line in lines:
  new_string += line.rstrip()

# Priting only the first few characters
print(new_string[:50])
```

```python
# Writing
file_name = 'string.txt'

n = 3
to_write = 'Hi all. This is Python 2. \n'
# Opening the file
with open(file_name, 'w') as file_object:
  for value in range(n):
    file_object.write(to_write)


with open(file_name) as file_object:
  content = file_object.read()
  print(content)
```

```python
# Writing
file_name = 'user_input.txt'

print('Enter your user name')
print('\Enter q when finished')


while True:
   name = input('what is your name:  ')
   if name == "q":
      break
   else:
      print(f'hellow {name} welcome')
      with open(file_name, 'a') as file_object:
         file_object.write(f'{name} \n')
      with open(file_name) as file_object:
         content = file_object.read()
         print(content)
```
